```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = 'https://id.wikipedia.org/wiki/Daftar_film_Indonesia_terlaris_sepanjang_masa'
response = requests.get(url)
response
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
soup
```

```python
table = soup.find('table', class_='wikitable sortable')
```

```python
table_title = table.find_all('th')
table_title
```

```python
split_table_title = [title.text.strip() for title in table_title]
split_table_title
```

```python
df = pd.DataFrame(columns=split_table_title)
df
```

```python
table_row = table.find_all('tr')
table_row
```

```python
for row in table_row[1:]:
    try:
        row_data = row.find_all('td')
        individual_row_data = [data.text.strip() for data in row_data]
        length = len(df)
        df.loc[length] = individual_row_data
    except Exception:
        print('error in number',individual_row_data[0])
        continue 
```

```python
for row in table_row[1:]:

    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]
    print(individual_row_data)
```
